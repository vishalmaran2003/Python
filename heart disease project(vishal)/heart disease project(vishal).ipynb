```python
data["Age"].fillna(a, inplace=True)
data["Blood Pressure"].fillna(b, inplace=True)
data["Cholesterol Level"].fillna(c, inplace=True)
data["Alcohol Consumption"].fillna(d, inplace=True)
data["Sleep Hours"].fillna(e, inplace=True)
data["High Blood Pressure"].fillna(f, inplace=True)
data["High LDL Cholesterol"].fillna(g, inplace=True)
data["Stress Level"].fillna(h, inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
X=data.drop(["Exercise Habits","Family Heart Disease","BMI","Low HDL Cholesterol","Sugar Consumption","Triglyceride Level","Fasting Blood Sugar","CRP Level","Homocysteine Level","Heart Disease Status"],axis=1)
Y=data["Heart Disease Status"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
import joblib
joblib.dump(model,"heart_disease.pkl")
```
